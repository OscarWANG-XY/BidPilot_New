```python
import sys
from pathlib import Path
sys.path.append('..')

from docx_parser._01_xml_loader import DocxXMLLoader
from docx_parser._02_xml_parser import DocxXMLParser
from docx_parser._03_element_extractor import DocumentElementExtractor, ElementType
from docx_parser._04_structure_builder import DocumentTreeBuilder

# 1. 加载文档
#doc_path = "C:/Users/huiwa/Downloads/文本分析测试/CaseTest/case3_北京铁运投标人须知.docx"
#doc_path = "C:/Users/huiwa/Downloads/文本分析测试/CaseTest/case11-测试树结构.docx"
doc_path = "C:/Users/huiwa/Downloads/文本分析测试/CaseTest/case8：招标文件-第1包：一级压榨花生油.docx"
loader = DocxXMLLoader(doc_path)
raw_content = loader.extract_raw()
parser = DocxXMLParser(raw_content)

# 2. 提取元素
extractor = DocumentElementExtractor(parser)
elements = extractor.extract_all_elements()
```

```python
# 4. 基本统计
print(f"\n=== 基本统计 ===")
print(f"总元素数量: {len(elements)}")
print(type(elements))
print(elements[0],elements[1])
print(type(elements[0]),type(elements[1]))
from pprint import pprint
pprint(elements[0])
pprint(elements[1])
```

```python
from pprint import pprint
for element in elements:
    element_dict = element.__dict__
    element_dict.pop('raw_xml', None)
    pprint(element_dict)
```

```python
for element in elements:
    if element.element_type == ElementType.PARAGRAPH and element.is_heading:
        print(f"L{element.heading_level}: {element.content}")
```

```python
# 3. 构建文档树
tree_builder = DocumentTreeBuilder(elements)
doc_tree = tree_builder.build_tree()
```

```python
from pprint import pprint
print(type(doc_tree))
#pprint(doc_tree)
```

```python
# 4. 打印树结构
print("Document Tree Structure:")
print("----------------------")
tree_builder.print_tree()
```

```python
def print_first_level(root_node):
    print("\nFirst Level Content:")
    print("-------------------")
    for child in root_node.children:
        if child.element:
            print(f"- {child.element.content}")

# 使用方法
print_first_level(doc_tree)
```

```python
def print_first_level_with_siblings(root_node):
    print("\nFirst Level Content with Siblings:")
    print("----------------------------------")
    for child in root_node.children:
        if child.element:
            print(f"\nMain Node: {child.element.content}")
            # 打印前一个兄弟节点
            if child.prev_sibling and child.prev_sibling.element:
                print("Previous sibling:", child.prev_sibling.element.content)
            # 打印后一个兄弟节点
            if child.next_sibling and child.next_sibling.element:
                print("Next sibling:", child.next_sibling.element.content)

# 使用方法
print_first_level_with_siblings(doc_tree)
```

```python
def print_all_headings(root_node, level=0):
    """打印所有层级的标题和对应的路径"""
    if root_node.element and root_node.element.is_heading:
        indent = "  " * level
        print(f"{indent}[Path: {root_node.path}] {root_node.element.content}")
    
    for child in root_node.children:
        print_all_headings(child, level + 1)

# 使用方法
print("\nDocument Headings Structure:")
print("---------------------------")
print_all_headings(doc_tree)
```
