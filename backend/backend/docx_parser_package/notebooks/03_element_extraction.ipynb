```python
# 1. 加载文档
loader = DocxXMLLoader(doc_path)  # 创建DocxXMLLoader 加载器 实例 
raw_content = loader.extract_raw()  # 加载原始文档内容
parser = DocxXMLParser(raw_content)  # 创建DocxXMLParser 解析器 实例 
```

```python
# 2. 创建元素提取器
extractor = DocumentElementExtractor(parser)
```

```python
# 3. 提取所有元素
elements = extractor.extract_all_elements()
```

```python
# 4. 基本统计
print(f"\n=== 基本统计 ===")
print(f"总元素数量: {len(elements)}")
print(type(elements))
print(elements[0],elements[1])
print(type(elements[0]),type(elements[1]))
from pprint import pprint
pprint(elements[0])
pprint(elements[1])
```

```python
# 假设 elements 是已经提取的 DocumentElement 列表
toc_elements = [
    (elem.content, elem.toc_info['page_number']) 
    for elem in elements 
    if hasattr(elem, 'is_toc') and elem.is_toc
]

# 按顺序打印内容和页码
for content, page_number in toc_elements:
    print(f"{content} (页码: {page_number})")
```

```python
for element in elements:
    if element.element_type == ElementType.PARAGRAPH and element.is_heading:
        print(f"L{element.heading_level}: {element.content}")
```

```python
# 筛选出原来是标题但现在不是标题的元素
converted_headings = [
    elem for elem in elements 
    if elem.element_type == ElementType.PARAGRAPH 
    and elem.heading_info 
    and elem.heading_info.get('original_is_heading') 
    and not elem.is_heading
]

# 打印这些元素的信息
for elem in converted_headings:
    print(f"Content: {elem.content}")
    print(f"Original heading level: {elem.heading_info.get('original_heading_level')}")
    print(f"Original heading type: {elem.heading_info.get('original_heading_type')}")
    print("---")
```

```python
# 打印所有段落的第一行
for element in elements:
    if element.element_type == ElementType.PARAGRAPH:
        print(f"第一行: {element.content.strip().split('\n')[0]}")
```
